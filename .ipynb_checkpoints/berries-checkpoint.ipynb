### PRECIPITATION

```python
import requests
```

```python
#setting up dates variable
months = np.arange(1,13)
months = np.char.mod('%d', months)

n = len(months)
for i in range(n):
    if len(months[i]) == 1:
        months[i] = '0'+months[i]
    else:
        months[i] = months[i]
        
days = np.arange(1, 32)
days = np.char.mod('%d', days)
for i in range(n):
    if len(days[i]) == 1:
        days[i] = '0'+days[i]
    else:
        days[i] = days[i]
        
temp = [(x, y) for x in months for y in days]
dates = []
for tup in temp:
    dates.append(''.join(tup))
```

```python
# DOWNLOADING FILES

#2019 remove: 31 from 04, 06, 11 & 29,30,31 from 02
Remove19 = ('0431', '0631','0631','1131','0229','0230','0231')
dates2019 = [e for e in dates if e not in Remove19]

#2020 remove: 31 from 04, 06, 11 & 30,31 from 02
Remove20 = ('0431', '0631','0631','1131','0230','0231')
dates2020 = [e for e in dates if e not in Remove20]


2019
file_pfx = 'https://www.ncei.noaa.gov/data/precipitation-persiann/access/2019/PERSIANN-CDR_v01r01_2019'
file_sfx = '_c20190806.nc'

for date in dates2019:
    url2019 = file_pfx+date+file_sfx
    r = requests.get(url2019, allow_redirects = True)
    open('2019'+date+'.nc', 'wb').write(r.content)
    
    
2020
file_pfx = 'https://www.ncei.noaa.gov/data/precipitation-persiann/access/2020/PERSIANN-CDR_v01r01_2020'
file_sfx = '_c20200522.nc'

for date in dates2020:
    url2020 = file_pfx+date+file_sfx
    r = requests.get(url2020, allow_redirects = True)
    open('2020'+date+'.nc', 'wb').write(r.content)
```

### Reclassifying rasters

```python
#read in merged files
fert_path = 'C:/Users/rache/Downloads/data_science/berries/site_fert_mosaic.tif'
stand_path = 'C:/Users/rache/Downloads/data_science/berries/stand_diameter_mosaic.tif'
spruce_path = 'C:/Users/rache/Downloads/data_science/berries/spruce_volume_mosaic.tif'
site_fert_raster = rasterio.open(fert_path)
stand_diam_raster = rasterio.open(stand_path)
spruce_vol_raster = rasterio.open(spruce_path)

#convert to np array
site_fert_raster = site_fert_raster.read(1)
stand_diam_raster = stand_diam_raster.read(1)
spruce_vol_raster = spruce_vol_raster.read(1)
```

```python
#convert to floats
site_fert = site_fert_raster.astype('f4')
stand_diam = stand_diam_raster.astype('f4')
spruce_vol = spruce_vol_raster.astype('f4')
```

- class 1 is herb rich sites
- class 2 is herb rich heath forests
- class 3 is mesic forests
- class 4 is sub-xeric forests
- class 5 is xeric forests
- class 6 is barren forests 
- class 7 (rocky and sandy soils and alluvial lands) can be forest land, poorly productive forest land, or unproductive land
- class 8 (summit and fjeld land with single coniferous trees) is either poorly productive forest land or unproductive land
- class 9 (mountain birch dominated fjelds) is poorly productive forest land or unproductive land
- class 10 (Open fjelds) is poorly productive forest land or unproductive land


- 32766  pixel where result should have been computed but it was not possible because of, e.g., clouds
- 32767  pixel that is outside forest land, poorly productive land, or unproductive land, or is not in Finland

```python
#view categories
uniq_vals = np.unique(site_fert)
print(sorted(uniq_vals))
```

Reclass site fertility:
- classes 1-4: good
- classed 5-10: bad

```python
#reclassifying
site_fert[site_fert <= 4] = 2
site_fert[(5 <= site_fert) & (site_fert <= 11)] = 1
site_fert[site_fert >= 30000] = -1
```
